```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./../outputs/prefetchcount_1.csv')
df1 = pd.read_csv('./../outputs/prefetchcount_10.csv')
df2 = pd.read_csv('./../outputs/prefetchcount_1000.csv')
df['TIME'] = df.index
df1['TIME'] = df1.index
df2['TIME'] = df2.index
# Drop pc_computed column
df = df.drop(columns=['PC_COMPUTED'], axis=1)
df1 = df1.drop(columns=['PC_COMPUTED'], axis=1)
df2 = df2.drop(columns=['PC_COMPUTED'], axis=1)
```

```python
# Plot arrival rate vs time with legend and title
def plot_arrival_rate_vs_time(df, pc_value):
    plt.figure(figsize=(10, 5))
    plt.title('Arrival Rate vs Time for PC = ' + str(pc_value))
    plt.xlabel('Time in seconds')
    plt.ylabel('Arrival Rate')
    sns.set_style("darkgrid")
    sns.lineplot(x='TIME', y='ARRIVAL_RATE', data=df)
    plt.show()
```

```python
plot_arrival_rate_vs_time(df, 1)
```

```python
plot_arrival_rate_vs_time(df1, 10)
```

```python
plot_arrival_rate_vs_time(df2, 1000)
```
